```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model = "llama-3.1-8b-instant")
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content = "Hi, My name is Nidhi and I am a GenAI Engineer")])
```

```python
from langchain_core.messages import AIMessage
model.invoke([HumanMessage(content = "Hi, My name is Nidhi and I am a GenAI Engineer"),
AIMessage(content = "Nice to meet you, Nidhi. As a GenAI Engineer, I'm sure you're working on exciting projects that involve the development and application of artificial intelligence and machine learning technologies. \n\nCould you tell me a bit more about your work or what you're currently focusing on? I'm all ears and happy to chat about AI-related topics."),
HumanMessage(content = "Hi, What is my name and What do I do")])
```
